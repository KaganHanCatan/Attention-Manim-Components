```python
%%manim -qm -v WARNING SingleHeadSelfAttention

from manim import *

class SingleHeadSelfAttention(Scene):
    def construct(self):
        # Create a square for Q
        square_q = Square(side_length=3, color=BLACK)
        letter_q = MathTex("Q").move_to(square_q.get_center())
        dimension_text_q = MathTex("(6,512)").next_to(letter_q, DOWN * 2.5)

        # Create the multiplication symbol
        x = MathTex("\\times").next_to(square_q, RIGHT * 2)

        # Create a square for K^T
        square_k = Square(side_length=3, color=BLACK).next_to(x, RIGHT)
        letter_k = MathTex("K^T").move_to(square_k.get_center())
        dimension_text_k = MathTex("(6,512)").next_to(letter_k, DOWN * 2.5)

        # Animate the elements
        self.play(Write(square_q))
        self.play(Write(letter_q))
        self.play(Write(dimension_text_q))

        self.play(Write(x))

        self.play(Create(square_k))
        self.play(Write(letter_k))
        self.play(Write(dimension_text_k))

        # Wait for a moment to display the final scene
        self.wait(2)
```

```python
%%manim -qm -v WARNING SingleHeadSelfAttention

from manim import *

class SingleHeadSelfAttention(Scene):
    def construct(self):
        # Create a square
        square_q = Square(side_length=3, color=BLACK)

        # Create the letter 'Q' in the center
        letter_q = MathTex("Q").move_to(square_q.get_center())

        # Create dimension label
        dimension_text = MathTex("(6,512)").next_to(letter_q, DOWN * 2.5)

        x = MathTex("\\times").next_to(square_q, RIGHT * 2)

        square_k = Square(side_length=3, color=BLACK).next_to(x, RIGHT)
        letter_k = MathTex("K^T").move_to(square_k.get_center())
        dimension_text_k = MathTex("(512,6)").next_to(letter_k, DOWN * 2.5)

        # Create division line
        division_line = Line(start=LEFT * 4, end=RIGHT * 4).next_to(x, DOWN * 7).shift(LEFT * 0.2)

        denominator = MathTex("\\sqrt{512}").next_to(division_line, DOWN * 1.5)

        left_p = MathTex("\Biggl(").next_to(division_line, LEFT * 1.5).shift(UP * 1)
        left_p.scale(2.5)
        right_p = MathTex("\Biggl)").next_to(division_line, RIGHT * 1).shift(UP * 1)
        right_p.scale(2.5)
        softmax = MathTex("\\text{softmax}").next_to(left_p, LEFT * 1)

        equal_sign = MathTex("=").next_to(right_p, RIGHT * 1)

        all_components = VGroup(square_q, dimension_text, letter_q, x, square_k, letter_k, dimension_text_k, division_line, denominator, left_p, right_p, softmax, equal_sign).move_to(LEFT * 1)
        all_components.scale(0.5)

        # Table data (without labels)
        table_data = [
            [".", "I", "love", "my", "cat", "very", "much"],
            ["I", "0.2", "0.5", "0","0", "0", "0.1"],
            ["love", "0.1", "0.3", "0","0", "0", "0.6"],
            ["my", "0.7", "0.2", "0", "0", "0", "0.3"],
            ["cat", "0.3", "0.8", "0", "0", "0", "0.2"],
            ["very", "0.9", "0.1", "0", "0", "0", "0.5"],
            ["much", "0.2", "0.6", "0", "0", "0", "0.4"],
        ]

        # Create the table
        table = Table(
            table_data,
            include_outer_lines=True,
            line_config={"stroke_width": 2, "color": BLACK},
            element_to_mobject_config={"font_size": 24},
        )
        table.scale(0.3)  # Adjust scale as needed
        table.next_to(all_components, RIGHT * 0.5)

        # Color the first column orange
        for i in range(len(table_data)):
            table.add_highlighted_cell((i + 1, 1), color=WHITE)

        # Color the rest of the table green
        for i in range(len(table_data)):
            for j in range(2, len(table_data[0]) + 1):
                table.add_highlighted_cell((i + 1, j), color=WHITE)

        # Create a brace above the table
        brace = Brace(table.get_columns(), direction=UP)
        brace_text = Text("Kelimeler Arasndaki ilikinin iddeti", font_size=10).next_to(brace, UP)
        under_table_text = MathTex(r"(6, 6)").scale(0.3).next_to(table, DOWN)

        # Add everything to the scene
        self.add(all_components, table, brace, brace_text, under_table_text)
```

```python
%%manim -qm -v WARNING MultiHeadAttentionFormula

from manim import *

class MultiHeadAttentionFormula(Scene):
    def construct(self):
        # Single-Head Self-Attention Formula
        attention = MathTex(
            "\\text{Attention}(Q, K, V) = \\text{softmax}\\left(\\frac{QK^T}{\\sqrt{d_k}}\\right) V"
        ).scale(0.8).move_to(ORIGIN)

        multi_head = MathTex(
            "\\text{MultiHead}(Q,K,V) = \\text{Concat}(head_1, \\cdots, head_h) W^0"
        ).scale(0.8).next_to(attention, DOWN * 2.5)

        head = MathTex(
            "\\text{head}_i = \\text{Attention}(QW_i^Q, KW_i^K, VW_i^V)"
        ).scale(0.8).next_to(multi_head, DOWN * 2.5)

        group = VGroup(attention, multi_head, head).move_to(ORIGIN)
        self.add(group)
```
